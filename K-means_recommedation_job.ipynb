# Lọc stopword

```python
import os
import re

import nltk
from emot.emo_unicode import EMOTICONS_EMO, UNICODE_EMOJI
from nltk.stem import WordNetLemmatizer

nltk.download("wordnet")
nltk.download("stopwords")
stop_words = set(nltk.corpus.stopwords.words("english"))



def delete_numbers(text):
    return "".join([c for c in text if not c.isdigit()])



def decode_abbreviations(text):
    standardizing_dict = {}
    with open("en-abbreviations.txt", "r", encoding="utf-8") as f:
        for row in f:
            row = row.strip().replace("\t\t\t", "\t").replace("\t\t", "\t")
            parts = row.split("\t")
            if len(parts) < 2:
                continue
            acronym, word = parts[:2]
            acronym = acronym.replace(" ", "")
            word = word.split(",")[0].replace(" ", "")
            standardizing_dict[acronym] = word
    return text

def handle_emojis(text):
    for x in EMOTICONS_EMO:
        text = text.replace(
            x,
            "_".join(EMOTICONS_EMO[x].replace(",", "").replace(":", "").split()),
        )
    for x in UNICODE_EMOJI:
        text = text.replace(
            x,
            "_".join(UNICODE_EMOJI[x].replace(",", "").replace(":", "").split()),
        )
    return text

def tokenize_text(text):
    return re.findall(r"\b\w+\b", text)

def lemmatization(words):
    lemmatizer = WordNetLemmatizer()
    return [lemmatizer.lemmatize(word, "v") for word in words]
def remove_stopword(words):
    return [word for word in words if word not in stop_words]

def text_preprocessing(text):
    text = delete_numbers(text)
    text = decode_abbreviations(text)
    text = handle_emojis(text)
    words = tokenize_text(text)
    words = lemmatization(words)
    words = remove_stopword(words)
    return " ".join(words)

data['Content'] = data['Content'].apply(text_preprocessing)


```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
def preprocessing(text):
    if isinstance(text, list):
        text = ' '.join(text)
    text = text.replace("\n", "")  # Remove the newline command
    text = re.sub(r'\W', ' ', text)  # Remove all non-word characters except spaces
    text = text.lower()  # Convert text to lowercase
    return text
```

```python
documents = [' '.join(map(str, job)) for job in data[["Job Title", "Description", "Required Skills", 'Content']].values]
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(documents)
tfidf_matrix.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
from sklearn.metrics.pairwise import cosine_similarity

# Preprocess the resume text
resume_text = preprocessing(resume_text)

# Transform the resume text using the fitted vectorizer
resume_vector = tfidf_vectorizer.transform([resume_text])

# Calculate cosine similarities
cosine_similarities = cosine_similarity(resume_vector, tfidf_matrix)

# Select the top 10 most similar jobs
top_10_job_indices = cosine_similarities.argsort(axis=1)[:, -10:][0][::-1]
top_10_job_info = [data.iloc[i] for i in top_10_job_indices]
top_10_job_df = pd.DataFrame(top_10_job_info)

# # Optionally, drop the 'Content' column if not needed
# top_10_job_df = top_10_job_df.drop(columns=['Content'], axis=1)
top_10_job_df
```

## Gap- statics để lựa cụm K tối ưu

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist

def optimalK(data, nrefs=3, maxClusters=20):
    gaps = np.zeros((len(range(1, maxClusters+1)),))
    result_list = []
    for gap_index, k in enumerate(range(1, maxClusters+1)):
        km = KMeans(n_clusters=k)
        km.fit(data)
        disp = sum(np.min(cdist(data, km.cluster_centers_, 'euclidean'), axis=1)) / data.shape[0]
        refDisps = np.zeros(nrefs)
        for i in range(nrefs):
            randomReference = np.random.random_sample(size=data.shape)
            km = KMeans(n_clusters=k)
            km.fit(randomReference)
            refDisp = sum(np.min(cdist(randomReference, km.cluster_centers_, 'euclidean'), axis=1)) / randomReference.shape[0]
            refDisps[i] = refDisp
        gap = np.log(np.mean(refDisps)) - np.log(disp)
        gaps[gap_index] = gap
        result_list.append({'clusterCount':k, 'gap':gap})
    resultsdf = pd.DataFrame(result_list)
    return (gaps.argmax() + 1, resultsdf)

# Sử dụng hàm
optimalK_value, gap_df = optimalK(tfidf_svd, nrefs=5, maxClusters=10)
print('K tối ưu theo Gap Statistics:', optimalK_value)


```

## Dùng phương pháo elbow để tìm ra K cụm tối ưu =>> dùng K-means để phâm cụm

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Áp dụng phương pháp Elbow
sum_of_squared_distances = []
K = range(1, 11)
for k in K:
    km = KMeans(n_clusters=k,init='k-means++', max_iter=300, n_init=10, random_state=42)
    km = km.fit(tfidf_svd)
    sum_of_squared_distances.append(km.inertia_)
```

```python
# Vẽ đồ thị để tìm 'Elbow'
plt.plot(K, sum_of_squared_distances, 'bx-')
plt.xlabel('K')
plt.ylabel('Tổng bình phương khoảng cách')
plt.title('Phương pháp Elbow cho giá trị K tối ưu')
plt.show()
```

Với elbow, ta nhận thấy khá khó tìm khuỷa tay =>>> chuyển qua silhouette

## Shihoutle để xác định cụm tối ưu 

```python
from sklearn.metrics import silhouette_score
```

```python
# Khởi tạo danh sách để lưu điểm silhouette
silhouette_scores = []

# Lặp qua các giá trị K và tính điểm silhouette
for k in range(2, 11):  # silhouette score không thể tính với k=1
    km = KMeans(n_clusters=k, init='k-means++', max_iter=300, n_init=10, random_state=42)
    km.fit(tfidf_svd)
    labels = km.labels_
    score = silhouette_score(tfidf_svd, labels)
    silhouette_scores.append(score)
```

```python
# Vẽ đồ thị cho điểm silhouette
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(range(2, 11), silhouette_scores, 'bo-')
plt.xlabel('K')
plt.ylabel('Điểm Silhouette')
plt.title('Phương pháp Silhouette cho giá trị K tối ưu')
```

K=2 là tối ưu
